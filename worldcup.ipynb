```python
# get the player details with highest runs given and the oppenent
player_name=df.loc[df['runs'].idxmax(),'player']
team_name=df.loc[df['runs'].idxmax(),'team']
highest_runs=df.loc[df['runs'].idxmax(),'runs']
opponent=df.loc[df['runs'].idxmax(),'opponent']

print('highest runs given by',player_name,'from',team_name,'whose runs given are',highest_runs,'against',opponent)

## get the player details with highest wickets taken and the oppenent
player_name=df.loc[df['wickets'].idxmax(),'player']
team_name=df.loc[df['wickets'].idxmax(),'team']
wickets=df.loc[df['wickets'].idxmax(),'wickets']
opponent=df.loc[df['wickets'].idxmax(),'opponent']

print('highest wickets taken by',player_name,'from',team_name,'whose wickets  are',wickets,'against',opponent)

#6s conceded by which player and against which country
player_name=df.loc[df['6s'].idxmax(),'player']
team_name=df.loc[df['6s'].idxmax(),'team']
highest_6s=df.loc[df['6s'].idxmax(),'6s']
opponent=df.loc[df['6s'].idxmax(),'opponent']

print('highest 6s given by',player_name,'from',team_name,'whose conceded',highest_6s,'against',opponent)

#4s conceded
player_name=df.loc[df['4s'].idxmax(),'player']
team_name=df.loc[df['4s'].idxmax(),'team']
highest_4s=df.loc[df['4s'].idxmax(),'4s']
opponent=df.loc[df['4s'].idxmax(),'opponent']
print('highest 4s given by',player_name,'from',team_name,'whose conceded',highest_4s,'against',opponent)

#most maidens
player_name=df.loc[df['maidens'].idxmax(),'player']
team_name=df.loc[df['maidens'].idxmax(),'team']
highest_maidens=df.loc[df['maidens'].idxmax(),'4s']
opponent=df.loc[df['maidens'].idxmax(),'opponent']
print('highest maidens by',player_name,'from',team_name,'who had',highest_maidens,'against',opponent)

```

```python
#group players by bowling stats
df_player_overs=df.groupby('player').agg({'overs':'sum','maidens':'sum','wickets':'sum','runs':'sum','run_rate':'mean','6s':'sum','4s':'sum','maidens':'sum','wd':'sum'})
```

```python
sort_bowler_stats=df_player_overs.sort_values(by='overs',ascending=False).reset_index()
```

```python
sort_bowler_stats.rename(columns={'wd':'wides'},inplace=True)
```

```python
sort_bowler_stats
```

```python
#Plotting the overs bowled by top 20bowlers
top20bowlers=sort_bowler_stats.head(20)
plt.figure(figsize=(15,6))
plt.title('Top 20 bowlers',fontsize=20)
plt.bar(top20bowlers['player'],top20bowlers['overs']) #x=players,y=overs
plt.xlabel('Player')
plt.xticks(rotation=45,ha='right',fontsize=10)#rotate the names and resize the font
plt.ylabel('Overs Bowled')
plt.tight_layout() #ensures the values are fit in figure area
plt.show()
```

```python
#runs given by top 20 bowlers
top20bowlers=sort_bowler_stats.head(20)
plt.figure(figsize=(15,6))
plt.title('Runs Given',fontsize=20)
plt.bar(top20bowlers['player'],top20bowlers['runs'])
plt.xlabel('Player')
plt.xticks(rotation=45,ha='right',fontsize=10) #rotate the names
plt.ylabel('Runs Given')
plt.tight_layout()
plt.show()
```

```python
#run rate for th etop 20 bowlers
plt.figure(figsize=(15,6))
plt.bar(top20bowlers['player'],top20bowlers['run_rate'])
plt.xlabel('Player')
plt.ylabel('Run Rate')
plt.title('Run rate of top 20 bowlers',fontsize=20)
plt.xticks(rotation=45,ha='right',fontsize=10)
plt.tight_layout()
plt.show()
```

```python
#wide balls from the top 20 bowlers
plt.figure(figsize=(15,6))
plt.title('Wides',fontsize=20)
plt.bar(top20bowlers['player'],top20bowlers['wides'])
plt.xlabel('Player',fontsize=10)
plt.ylabel('Wides Bowled')
plt.xticks(rotation=45,ha='right',fontsize=14)
plt.tight_layout()
plt.show()
```

```python
#plotting a subplot for overs bowled and maidens
#This creates a figure (fig) and a set of subplots (ax). ax is a single subplot, and fig is the entire figure object.
fig,ax =plt.subplots(figsize=(12,6))

#plot bars for y axis
index=np.arange(len(top20bowlers))
bars=ax.bar(index,top20bowlers['overs'],label='Overs',color='skyblue')

#plot line for y axis
line=ax.plot(index,top20bowlers['maidens'],color='orange',label='Maidens',marker='o')

#set title and label
ax.set_xlabel('Player')
ax.set_ylabel('Overs',color='skyblue')
plt.title('Maiden overs and overall overs bowled')

#set x axis and label
plt.xticks(index,top20bowlers['player'],label='Players',ha='right',rotation=45)

#legends
bar_legend=[bar for bar in bars]
line_legend=[line[0]]

plt.legend(bar_legend+line_legend,['Overs','Maiden'])

plt.tight_layout()
plt.show()
```

```python
sort_bowler_wickets=df_player_overs.sort_values(by='wickets',ascending=False).reset_index()

top20wickets=sort_bowler_wickets.head(20)
# plot wickets by top20 players
plt.figure(figsize=(15,6))
plt.title('Top Wickets taken')

plt.bar(top20wickets['player'],top20wickets['wickets'])

plt.xlabel('Players',fontsize=20)
plt.ylabel('Wickets')
plt.xticks(label='Players',ha='right',rotation=45)
plt.yticks(label='Wickets')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


# Create evenly spaced x-axis positions for the grouped bars
x_positions = np.arange(len(top20wickets['player']))
bar_width = 0.35  # Width of each bar

# Plot wickets and overs for top 20 players
plt.figure(figsize=(15, 7))
plt.title('Top Wickets and Overs Taken')

plt.bar(x_positions - bar_width/2, top20wickets['wickets'], width=bar_width, label='Wickets')
plt.bar(x_positions + bar_width/2, top20wickets['overs'], width=bar_width, label='Overs', color='orange')

plt.xlabel('Players', fontsize=20)
plt.ylabel('Overs and wickets',fontsize=20)
plt.xticks(x_positions, top20wickets['player'], rotation=45, ha='right',fontsize=15)
plt.legend()  # Add legend to distinguish between bars
plt.tight_layout()
plt.show()
```
